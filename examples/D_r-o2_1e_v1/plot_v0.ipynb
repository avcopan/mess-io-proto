```python
from pathlib import Path

import pyparsing as pp

mess_inp_file = Path("01_1-30_mess.inp")
spc_file = Path("species.csv")
```

```python
from mess_io_proto import surface

surf = surface.from_mess(mess_inp_file, spc_inp=spc_file)
surf = surface.with_energies_relative_to(surf, 23)
surf = surface.without_fake_wells(surf)
surface.display_network(surf)
```

```python
from matplotlib import pyplot

size = (10, 5)
fig = pyplot.figure(figsize=size)
fig = pyplot.figure()


class Color:
    black = "#000000"
    gray1 = "#808080FF"
    gray2 = "#BFBFBFFF"
    red = "#FF2E2EFF"
    blue = "#4D4DFFFF"
    blue1 = "#9999ffff"
    green = "#31FF31FF"
    green1 = "#99ff99ff"
    yellow = "#ffcd00ff"


paths, colors = zip(
    ([23, 11, 1, 19], Color.black),
    ([22, 0, 1, 19], Color.gray1),
    ([22, 0, 4, 19], Color.gray2),
    ([22, 0, 1, 10], Color.red),
    ([23, 11, 18], Color.blue),
    ([22, 0, 18], Color.blue1),
    ([23, 11, 1, 18], Color.green),
    ([22, 0, 4, 18], Color.green1),
    ([22, 0, 1, 21], Color.yellow),
    strict=True,
)

fig = surface.plot_paths(
    surf,
    paths=paths,
    fig=fig,
    colors=colors,
    stereo=True,
)

pyplot.savefig("plot.svg")
```
